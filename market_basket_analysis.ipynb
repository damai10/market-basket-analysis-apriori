# 🛒 Market Basket Analysis using Apriori Algorithm

## 📌 Objective
The goal of this project is to identify relationships between products frequently purchased together using transaction-level data.

Market Basket Analysis helps businesses understand customer buying behavior and uncover cross-selling opportunities.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

## 📊 Dataset Overview

The dataset contains retail transaction data including:
- InvoiceNo (Transaction ID)
- Description (Product Name)
- Quantity (Number of items purchased)
- InvoiceDate (Transaction Date)
- CustomerID

Each row represents a product purchased within a transaction.

```python
df = pd.read_excel("Online+Retail.xlsx")
df.head()
```

## 🧹 Data Cleaning

Before analysis, the dataset was cleaned by:
- Removing missing product descriptions
- Filtering out negative quantities (returns)
- Ensuring correct data types for processing

This step ensures accurate and reliable results.

```python
# Remove nulls
df = df.dropna(subset=['Description'])
```

```python
# Remove negative/returns
df = df[df['Quantity'] > 0]
```

```python
# Convert to string
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
```

```python
basket = (df
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0))
```

```python
def encode(x):
    if x > 0:
        return 1
    return 0

basket = (basket > 0).astype(int)
```

```python
frequent_items = apriori(basket, min_support=0.02, use_colnames=True)
```

```python
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
rules.head()
```

```python
rules.sort_values(by='lift', ascending=False).head(10)
```
